```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
L = 50 # supremum of the sigmoid function
sigmoid = lambda x: 1 / (1 + np.exp(-x))
logistic = lambda x: L * sigmoid(x)
x = np.linspace(-10, 10)
y = logistic(x)

plt.plot(x, y, label=f"$\\sigma(x)=\\frac{{{L}}}{{1 + e^{{-x}}}}$")
plt.axhline(0, linestyle='--', color="gray")
plt.axhline(L / 2, linestyle='--', color="gray")
plt.axhline(L, linestyle='--', color="gray")
plt.annotate("0", (7.5, 1))
plt.annotate(str(L / 2), (7.5, L / 2 + 1))
plt.annotate(str(L), (7.5, L - 2))
plt.xlabel("x")
plt.ylabel("$\\sigma(x)$")
plt.legend()
plt.show()
```

```python
omega = 0.4
omega_1 = omega
omega_2 = omega
I = -10

def det(x, y):
    d = np.zeros((len(x), len(y)))
    for i, xi in enumerate(x):
        for j, yi in enumerate(y):
            sigma_y = sigmoid(omega_2 * xi + I)
            sigma_x = sigmoid(omega_1 * yi + I)
            d[i,j] = 1 - 2500 * sigma_x * sigma_y * (1 - sigma_x) * (1 - sigma_y)
    return d

x = np.linspace(-10, 10, num=1000)
y = np.linspace(-10, 10, num=1000)
z = det(x, y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x, y, z, vmin=5)
plt.show()
```
